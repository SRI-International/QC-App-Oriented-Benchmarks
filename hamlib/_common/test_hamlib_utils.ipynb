```python

from hamlib._common import hamlib_utils

hamlib_utils.verbose = False

####################

hamlib_utils.load_hamlib_file("FH_D-1")

datasets = hamlib_utils.find_dataset_for_params(12, { "graph-1D-grid": "pbc", "enc": "bk", "U":4 })

for dataset_name, dataset in datasets.items():
    print(f"dataset_name = {dataset_name}")
    print(f"attributes = {dataset['attributes']}")
    print(f"data = {dataset['data']}")
    print("")
    
# return a sparse Pauli list of terms queried from the open HamLib file
sparse_pauli_terms, dataset_name = hamlib_utils.get_hamlib_sparsepaulilist(num_qubits=12,
                                        params={ "graph-1D-grid": "pbc", "enc": "bk", "U":4 })
print(sparse_pauli_terms)
```
