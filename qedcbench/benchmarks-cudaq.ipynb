Configure the first cell with your backend and execution settings, then run the benchmark cells of interest.

# Execute Benchmarks with Noise
The next set of cells configures the benchmarks to execute on the density-matrix-cpu simulator where noise is simulated during execution.
Noisy simulation is costly and hence we limit the number of qubits executed to only 10 here.
The default noise model used sets a depolarization channel with a value 0.04 on all single-qubit operations.

```python
# Noisy simulation settings (override app_args and exec_args for noisy execution)
app_args = dict(
    min_qubits = 2,
    max_qubits = 8,                     # execute fewer qubits with noise model
    skip_qubits = 1,
    max_circuits = 1,
    num_shots = 1000,
    api = "cudaq",
)

exec_args = dict(
    backend_id = "density-matrix-cpu",  # noisy simulation target
    hub = "", group = "", project = "",
    provider_backend = None,
    exec_options = {"noise_model": "default"},
)

# Add a suffix to data/image files to distinguish noisy results
from qedclib import metrics
metrics.data_suffix = "-mysystem-noisy"
```

### Bernstein-Vazirani - Method 1

```python
from bernstein_vazirani import bv_benchmark
bv_benchmark.run(method=1, **app_args, **exec_args)
```

### Hidden Shift

```python
from hidden_shift import hs_benchmark
hs_benchmark.run(**app_args, **exec_args)
```

### Quantum Fourier Transform - Method 1

```python
from quantum_fourier_transform import qft_benchmark
qft_benchmark.run(method=1, **app_args, **exec_args)
```

### Quantum Phase Estimation

```python
from phase_estimation import pe_benchmark
pe_benchmark.run(**app_args, **exec_args)
```
